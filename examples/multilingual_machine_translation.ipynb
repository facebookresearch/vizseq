## 1. Set up data inputs

````python
from glob import glob
root = 'data/multilingual_translation_ted58_x_en_test'
src, ref, hypo = sorted(glob(f'{root}/src_*.*')), glob(f'{root}/ref_*.txt'), glob(f'{root}/pred_*.txt')
tag = glob(f'{root}/tag_*.txt')
````

## 2. Analysis with VizSeq

````python
import vizseq
````

````python
vizseq.view_stats(src, ref)
````

````python
vizseq.view_n_grams(src, k=10)
````

````python
vizseq.view_scores(ref, hypo, ['bleu'], tags=tag)
````

````python
vizseq.view_examples(src, ref, hypo, ['bleu'], page_sz=2, page_no=16)
````
